```python
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
mask1 = skimage.io.imread('output/brain/125-25/final/0020.png_learned_mask.jpg')
# unique, counts = np.unique(mask1, return_counts=True)
# dict(zip(unique, counts))
# mask1.shape
```

```python
mask2 = skimage.io.imread('output/resume/2/final/mask_3800.jpg')
```

```python
figure, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(mask1, cmap='gray')
ax[1].imshow(mask2, cmap='gray')
```

```python
def combine_OR(m1, m2):
    m3 = np.zeros_like(m1)
    m4 = np.zeros_like(m2)
    m3[m1 >= 128] = 1
    m3[m1 < 128] = 0
    m4[m2 >= 128] = 1
    m4[m2 < 128] = 0
    m5 = m3 + m4
    return m5
```

```python
mask3 = combine_OR(mask1, mask2)
```

```python
plt.imshow(mask3, cmap='gray')
```

### Better Hints
why fg_hint, bg_hint not 255 reverse relation?

```python
img = skimage.io.imread('data/FISH/images/0000.png')
unique, counts = np.unique(img, return_counts=True)
# dict(zip(unique, counts))
```

```python
t1 = 95
t2 = 105

img_fg = img.copy()
img_fg[img > t1] = 255
img_fg[img <= t1] = 0

img_bg = img.copy()
img_bg[img >= t2] = 0
img_bg[img < t2] = 255
```

```python
figure, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(img)
ax[1].imshow(img_fg)
ax[2].imshow(img_bg)
```
